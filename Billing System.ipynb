```python
class Bill_App:
    def __init__(self, root):
        self.root = root
        self.root.geometry("1530x800+0+0")
        self.root.title("AD Billing Software")
        

        lbltitle=Label(self.root,bd=20,relief=RIDGE,text="AD Mini Mall Billing",fg="#DC143C",bg="#CDAA7D",font=("Arial",50,"bold"))
        lbltitle.pack(side=TOP,fill=X)

        #Variable=========================================================================================

        self.Customer_name=StringVar()
        self.Mobile_no=StringVar()
        self.Email=StringVar()
        self.Bill_no=StringVar()
        z=random.randint(1000,9999)
        self.Bill_no.set(z)
        self.Date=StringVar()
        self.Product_Id=StringVar()
        self.Product_name=StringVar()
        self.Category=StringVar()
        self.Price=IntVar()
        self.Quantity=IntVar()
        self.Tax=StringVar()
        self.Subtotal=StringVar()
        self.Total_Amount=StringVar()
        self.Search_bill=StringVar()

        



        
        
        #Main Frame===============================================================================
        
        Main_Frame=Frame(self.root,bd=5,relief=GROOVE,bg="#CDAA7D")
        Main_Frame.place(x=0,y=150,width=1580,height=620)

        #Midlle frame==============================================================================
        
        Middle_Frame=Frame(Main_Frame,bd=20,relief=RIDGE,bg="#8B7355")
        Middle_Frame.place(x=10,y=150,width=1010,height=340)
        


        # Customer Label Frame=======================================================================
        
        Cust_Frame=LabelFrame(Main_Frame,text="Customer",font=("Arial",12,"bold"),bg="white",fg="#8B3E2F")
        Cust_Frame.place(x=10,y=5,width=350,height=140)
        
        #Mobile number===============================================================================
        
        self.lbl_mob=Label(Cust_Frame,text="Mobile No.",font=("Arial",12,"bold"),bg="white",fg="#8B3E2F")
        self.lbl_mob.grid(row=0,column=0,sticky=W,padx=5,pady=2)
        
        self.entry_mob=ttk.Entry(Cust_Frame,textvariable=self.Mobile_no,font=("Arial",10,"bold"),width=24)
        self.entry_mob.grid(row=0,column=1,sticky=W,padx=5,pady=2)
        
        #Customer Name===========================================================================
        
        self.lbl_cust=Label(Cust_Frame,text="Customer Name",font=("Arial",12,"bold"),bg="white",fg="#8B3E2F",bd=4)
        self.lbl_cust.grid(row=1,column=0,sticky=W,padx=5,pady=2)
        
        self.entry_cust=ttk.Entry(Cust_Frame,textvariable=self.Customer_name,font=("Arial",10,"bold"),width=24)
        self.entry_cust.grid(row=1,column=1,sticky=W,padx=5,pady=2)
        
        # Email===========================================================================
        
        self.lbl_email=Label(Cust_Frame,text="Email",font=("Arial",12,"bold"),bg="white",fg="#8B3E2F",bd=4)
        self.lbl_email.grid(row=2,column=0,sticky=W,padx=5,pady=2)
        
        self.entry_email=ttk.Entry(Cust_Frame,textvariable=self.Email,font=("Arial",10,"bold"),width=24)
        self.entry_email.grid(row=2,column=1,sticky=W,padx=5,pady=2)


        # Product Information Frame==============================================================
        
        Prod_Frame=LabelFrame(Main_Frame,text="Production Information",font=("Arial",12,"bold"),bg="white",fg="#008B8B")
        Prod_Frame.place(x=370,y=5,width=645,height=140)
        
        #Category==================================================================================
        
        self.lbl_category=Label(Prod_Frame,text="Product Id",font=("Arial",12,"bold"),bg="white",fg="#008B8B",bd=4)
        self.lbl_category.grid(row=0,column=0,sticky=W,padx=5,pady=2)
        
        self.Combo_Category=ttk.Entry(Prod_Frame,textvariable=self.Product_Id,font=("Arial",10,"bold"),width=24)
        self.Combo_Category.grid(row=0,column=1,sticky=W,padx=5,pady=2)
        self.Combo_Category.bind("<Return>", self.product_info)
        
        
        #Sub-Category================================================================================
        
        self.lbl_subcategory=Label(Prod_Frame,text="Category",font=("Arial",12,"bold"),bg="white",fg="#008B8B",bd=4)
        self.lbl_subcategory.grid(row=1,column=0,sticky=W,padx=5,pady=2)
        
        self.Combo_subCategory=ttk.Entry(Prod_Frame,textvariable=self.Category,font=("Arial",10,"bold"),width=24)
        self.Combo_subCategory.grid(row=1,column=1,sticky=W,padx=5,pady=2)
        
        
        #Product Name==================================================================================
        
        self.lbl_productname=Label(Prod_Frame,text="Product Name",font=("Arial",12,"bold"),bg="white",fg="#008B8B",bd=4)
        self.lbl_productname.grid(row=2,column=0,sticky=W,padx=5,pady=2)
        
        self.Combo_productname=ttk.Entry(Prod_Frame,textvariable=self.Product_name,font=("Arial",10,"bold"),width=24)
        self.Combo_productname.grid(row=2,column=1,sticky=W,padx=5,pady=2)
        
        #Price=======================================================================================
        
        self.lbl_price=Label(Prod_Frame,text="Price",font=("Arial",12,"bold"),bg="white",fg="#008B8B",bd=4)
        self.lbl_price.grid(row=0,column=2,sticky=W,padx=5,pady=2)
        
        self.Combo_price=ttk.Entry(Prod_Frame,textvariable=self.Price,font=("Arial",10,"bold"),width=24)
        self.Combo_price.grid(row=0,column=3,sticky=W,padx=5,pady=2)
        
        #Quantity========================================================================================
        
        self.lbl_qty=Label(Prod_Frame,text="Quantity",font=("Arial",12,"bold"),bg="white",fg="#008B8B",bd=4)
        self.lbl_qty.grid(row=1,column=2,sticky=W,padx=5,pady=2)
        
        self.entry_qty=ttk.Entry(Prod_Frame,textvariable=self.Quantity,font=("Arial",10,"bold"),width=26)
        self.entry_qty.grid(row=1,column=3,sticky=W,padx=5,pady=2)


         #Bill counter Label Frame=========================================================================
        
        Bottom_Frame=LabelFrame(Main_Frame,text="Bill Counter",font=("Arial",12,"bold"),bg="white",fg="#BDB76B")
        Bottom_Frame.place(x=0,y=485,width=1520,height=140)
        
        #Sub Total======================================================================================
        
        self.lbl_Subtotal=Label(Bottom_Frame,text="Sub Total",font=("Arial",12,"bold"),bg="white",bd=4)
        self.lbl_Subtotal.grid(row=0,column=0,sticky=W,padx=5,pady=2)
        
        self.entry_Subtotal=ttk.Entry(Bottom_Frame,textvariable=self.Subtotal,font=("Arial",10,"bold"),width=26)
        self.entry_Subtotal.grid(row=0,column=1,sticky=W,padx=5,pady=2)
        
        #TAX==========================================================================================
        
        self.lblTax=Label(Bottom_Frame,text="Govt Tax",font=("Arial",12,"bold"),bg="white",bd=4)
        self.lblTax.grid(row=1,column=0,sticky=W,padx=5,pady=2)
        
        self.entryTax=ttk.Entry(Bottom_Frame,textvariable=self.Tax,font=("Arial",10,"bold"),width=26)
        self.entryTax.grid(row=1,column=1,sticky=W,padx=5,pady=2)
        
        #Amount Total======================================================================================
        
        self.lblTotalAmount=Label(Bottom_Frame,text="Total Amount",font=("Arial",12,"bold"),bg="white",bd=4)
        self.lblTotalAmount.grid(row=2,column=0,sticky=W,padx=5,pady=2)
        
        self.entryTotalAmount=ttk.Entry(Bottom_Frame,textvariable=self.Total_Amount,font=("Arial",10,"bold"),width=26)
        self.entryTotalAmount.grid(row=2,column=1,sticky=W,padx=5,pady=2)

        # Bill Search=================================================================================

        Search_Frame=Frame(Main_Frame,bd=2,bg="white")
        Search_Frame.place(x=1030,y=12,width=500,height=40)
        
        self.lblBill=Label(Search_Frame,text="Bill Number",font=("Arial",12,"bold"),bg="#CDAA7D",fg="white")
        self.lblBill.grid(row=0,column=0,sticky=W,padx=1)
        
        self.entry_Search=ttk.Entry(Search_Frame,textvariable=self.Search_bill,font=("Arial",10,"bold"),width=26)
        self.entry_Search.grid(row=0,column=1,sticky=W,padx=2)
        
        
        self.BtnSearch=Button(Search_Frame,command=self.find_bill,text="Search",font=("Arial",11,"bold"),bg="#A2CD5A",fg="white",width=10,cursor="hand2")
        self.BtnSearch.grid(row=0,column=2)

        
        # Right Bill Area=======================================================================================
        
        RightLabelFrame=LabelFrame(Main_Frame,text="Bill Area",font=("Arial",12,"bold"),bg="white",fg="#CD3333")
        RightLabelFrame.place(x=1030,y=45,width=480,height=440)
        
        scroll_y=Scrollbar(RightLabelFrame,orient=VERTICAL)
        self.textarea=Text(RightLabelFrame,yscrollcommand=scroll_y.set,bg="white",fg="#FF6103",font=("Arial",12,"bold"))
        scroll_y.pack(side=RIGHT,fill=Y)
        scroll_y.config(command=self.textarea.yview)
        self.textarea.pack(fill=BOTH,expand=1)



         #Button Frame====================================================================================================
        
        Btn_Frame=Frame(Bottom_Frame,bd=2,bg="white")
        Btn_Frame.place(x=320,y=0)
        
        self.BtnAddtoCart=Button(Btn_Frame,command=self.Additem,height=2,text="Add to Cart",font=("Arial",15,"bold"),bg="#EE6A50",fg="white",width=12,cursor="hand2")
        self.BtnAddtoCart.grid(row=0,column=0)
        
        self.Btngenerate_bill=Button(Btn_Frame,command=self.gen_bill,height=2,text="Generate Bill",font=("Arial",15,"bold"),bg="#53868B",fg="white",width=12,cursor="hand2")
        self.Btngenerate_bill.grid(row=0,column=1)
        
        self.BtnSave=Button(Btn_Frame,command=self.save_bill,height=2,text="Save Bill",font=("Arial",15,"bold"),bg="#007FFF",fg="white",width=12,cursor="hand2")
        self.BtnSave.grid(row=0,column=2)
        
        self.BtnPrint=Button(Btn_Frame,command=self.print_bill,height=2,text="Print",font=("Arial",15,"bold"),bg="#EEAD0E",fg="white",width=12,cursor="hand2")
        self.BtnPrint.grid(row=0,column=3)
        
        self.BtnClear=Button(Btn_Frame,command=self.clear,height=2,text="Clear",font=("Arial",15,"bold"),bg="#EE7621",fg="white",width=12,cursor="hand2")
        self.BtnClear.grid(row=0,column=4)
        
        self.BtnExit=Button(Btn_Frame,command=self.root.destroy,height=2,text="Exit",font=("Arial",15,"bold"),bg="#3D59AB",fg="white",width=12,cursor="hand2")
        self.BtnExit.grid(row=0,column=5)

        self.BtnAddData = Button(
        Btn_Frame, height=2, text="Add Data", font=("Arial", 15, "bold"),bg="#DC143C", fg="white", width=12, cursor="hand2",command=self.iAddData)
        self.BtnAddData.grid(row=0, column=6)
        self.welcome()


        

    # =========================Scroll bar===========================================
        scroll_x=ttk.Scrollbar(Middle_Frame,orient=HORIZONTAL)
        scroll_y=ttk.Scrollbar(Middle_Frame,orient=VERTICAL)
        self.Category_table=ttk.Treeview(Middle_Frame,column=("Customer_name","Mobile_no","Email","Bill_no","Date","Product_Id","Product_name","Category","Price","Quantity","Subtotal","Tax","Total_Amount"),xscrollcommand=scroll_y.set,yscrollcommand=scroll_x.set)
        scroll_x.pack(side=BOTTOM,fill=X)
        scroll_y.pack(side=RIGHT,fill=Y)

        scroll_x=ttk.Scrollbar(command=self.Category_table.xview)
        scroll_y=ttk.Scrollbar(command=self.Category_table.yview)

        self.Category_table.heading("Customer_name",text="Name of Customer")
        self.Category_table.heading("Mobile_no",text="Mobile no.")
        self.Category_table.heading("Email",text="Email")
        self.Category_table.heading("Bill_no",text="Bill number")
        self.Category_table.heading("Date",text="Date")
        self.Category_table.heading("Product_Id",text="Product Id")
        self.Category_table.heading("Product_name",text="Product Name")
        self.Category_table.heading("Category",text="Category")
        self.Category_table.heading("Price",text="Price")
        self.Category_table.heading("Quantity",text="Quantity")
        self.Category_table.heading("Subtotal",text="Sub Total")
        self.Category_table.heading("Tax",text="Tax")
        self.Category_table.heading("Total_Amount",text="Total Amount")

        self.Category_table["show"]="headings"


        self.Category_table.column("Customer_name",width=120)
        self.Category_table.column("Mobile_no",width=120)
        self.Category_table.column("Email",width=100)
        self.Category_table.column("Bill_no",width=100)
        self.Category_table.column("Date",width=100)
        self.Category_table.column("Product_Id",width=100)
        self.Category_table.column("Product_name",width=100)
        self.Category_table.column("Category",width=100)
        self.Category_table.column("Price",width=100)
        self.Category_table.column("Quantity",width=100)
        self.Category_table.column("Subtotal",width=100)
        self.Category_table.column("Tax",width=100)
        self.Category_table.column("Total_Amount",width=100)    

        self.Category_table.pack(fill=BOTH,expand=1)
        self.l=[]
        
    def Additem(self):
        if self.Product_Id.get()=="":
            messagebox.showerror("Error","Please select product id")
        else:
            self.n=self.Price.get()
            self.m=self.Quantity.get()*self.n
            self.l.append(self.m)

            subtotal = sum(self.l)
            self.Subtotal.set(f'Rs. {subtotal:.2f}')

            tax_percentage = 12
            tax_amount = (subtotal - self.Price.get()) * tax_percentage / 100
            self.Tax.set(f'Rs. {tax_amount:.2f}')

            total_amount = subtotal + tax_amount
            self.Total_Amount.set(f'Rs. {total_amount:.2f}')

            self.textarea.insert(END,f"\n {self.Product_name.get()}\t\t\t{self.Quantity.get()}\t{self.Price.get()}\n")


        self.Product_Id.set("")
        self.Product_name.set("")
        self.Category.set("")
        self.Price.set(0)
        self.Quantity.set(1)
        
# Print BIll================================================================================================================
    
    def print_bill(self):
        q=self.textarea.get(1.0,"end-1c")
        filename=tempfile.mktemp('.txt')
        open(filename,'w').write(q)
        os.startfile(filename,"print")

# Save Bill ===============================================================================================================
    
    def save_bill(self):
        op = messagebox.askyesno("Save Bill", "Do you want to save the bill")
        if op > 0:
            if not hasattr(self, "bill_saved") or not self.bill_saved:
                self.bill_data = self.textarea.get(1.0, END)
    
                # Append customer details to the bill data only if they are not already present
                if not self.bill_data.startswith("Customer Name:"):
                    self.bill_data = (
                        f"Customer Name: {self.Customer_name.get()}\n"
                        f"Mobile No.: {self.Mobile_no.get()}\n"
                        f"Email: {self.Email.get()}\n"
                    ) + self.bill_data
    
                current_date_time = datetime.now().strftime("%Y-%m-%d %H-%M-%S")
                filename = os.path.join("D:/PythonPro/Bills", f"{self.Bill_no.get()}.txt")
                with open(filename, 'w') as file:
                    file.write(self.bill_data)
    
                # Clear the customer details from bill_data to prevent duplicate saves
                self.bill_data = ""
    
                # Set the flag to indicate that the bill has been saved
                self.bill_saved = True
    
                # Store the current date and time in the database
                self.insert_customer_details(current_date_time)
    
                messagebox.showinfo("Saved", f"Bill_no: {self.Bill_no.get()} saved successfully")
            else:
                messagebox.showinfo("Info", "Bill already saved")




    
# Generate Bill=====================================================================================================================
            
    def gen_bill(self):
        if not self.l:
            messagebox.showerror("Error", "Please add products to the cart")
        else:
              

            customer_details = (
                f"t\t\t  Welcome to AD Mall \n"
                f"Bill number:{self.Bill_no.get()}\n"
                f"Customer Name: {self.Customer_name.get()}\n"
                f"Mobile No.: {self.Mobile_no.get()}\n"
                f"Email: {self.Email.get()}\n"
            )
    
            
            current_bill = (
                "==================================================\n"
                + "Products:\t\t\tQTY\tPrice\n"
                + "==================================================\n"
                + self.textarea.get(10.0, (10.0 + float(len(self.l))))
            )
    
        
            complete_bill = (
                customer_details
                + "==================================================\n"
                + current_bill
                + f"\n Sub Total:\t\t\t{self.Subtotal.get()}\n"
                + f"Total Tax:\t\t\t{self.Tax.get()}\n"
                + f"Total Amount:\t\t\t{self.Total_Amount.get()}\n"
                + "==================================================\n"
            )
    
            
            self.textarea.delete(1.0, END)
            self.textarea.insert(END, complete_bill)
    
        
            self.Product_Id.set("")
            self.Product_name.set("")
            self.Category.set("")
            self.Price.set(0)
            self.Quantity.set(0)





# Find Bill==================================================================================================================
    
    def find_bill(self):
        found="no"
        for i in os.listdir("D:/PythonPro/Bills/"):
            if i.split('.')[0]==self.Search_bill.get():
                f1=open(f'D:/PythonPro/Bills/{i}','r')
                self.textarea.delete(1.0,END)
                for d in f1:
                    self.textarea.insert(END,d)
                f1.close()
                found="yes"
        if found=="no":
            messagebox.showerror("Error","Invalid bill number")


# Clear=======================================================================================================================
            
    def clear(self):
        self.textarea.delete(1.0,END)
        self.Customer_name.set("")
        self.Mobile_no.set("")
        self.Email.set("")
        self.Bill_no.set("")
        z=random.randint(1000,9999)
        self.Bill_no.set(z)
        self.Product_Id.set("")
        self.Product_name.set("")
        self.Category.set("")
        self.Price.set(0)
        self.Quantity.set(0)
        self.Tax.set("")
        self.Subtotal.set("")
        self.Total_Amount.set("")

        self.welcome()


# Textarea=====================================================================================================================    


    def welcome(self):
        self.textarea.delete(1.0, END)

        self.textarea.insert(END,"\t\t Welcome AD MALL")    
        self.textarea.insert(END, f"\n Bill Number:{self.Bill_no.get()}")
        self.textarea.insert(END, f"\n Customer Name:{self.Customer_name.get()}")
        self.textarea.insert(END, f"\n Mobile Number:{self.Mobile_no.get()}")
        self.textarea.insert(END, f"\n Email Id:{self.Email.get()}")
        self.textarea.insert(END, "\n==================================================")
        self.textarea.insert(END, f"\n Products:\t\t\tQTY\tPrice")
        self.textarea.insert(END, "\n==================================================\n")

        
    

#Functions=================================================================================================
    
    def iAddData(self):
        if self.Product_Id.get() == "":
            messagebox.showerror("Error", "Please enter correct product id")
        else:
            try:
                conn = mysql.connector.connect(
                    host="localhost",
                    user="root",
                    password="Kpml@11009",
                    database="python"
            )
                cursor = conn.cursor()
                query = "INSERT INTO Data (Product_Id, Product_name, Category, Price) VALUES (%s, %s, %s, %s)"
                values = (
                    self.Product_Id.get(),
                    self.Product_name.get(),
                    self.Category.get(),
                    self.Price.get()
            )

                cursor.execute(query, values)
                conn.commit()
                conn.close()
                messagebox.showinfo("Success", "Data inserted successfully")
            except Exception as e:
                messagebox.showerror("Error", str(e))

    

    def product_info(self, event):
        product_id = self.Product_Id.get()
        try:
            conn = mysql.connector.connect(
                host="localhost",
                user="root",
                password="Kpml@11009",
                database="python"
            )
            cursor = conn.cursor()
            
            query = "SELECT Product_name, Category, Price FROM Data WHERE Product_Id = %s"
            cursor.execute(query, (product_id,))
            result = cursor.fetchone()

            if result:
                product_name, category, price = result
                self.Product_name.set(product_name)
                self.Category.set(category)
                self.Price.set(price)
            else:
                messagebox.showerror("Error", "Product ID not found in the database")

            conn.close()
            
        except Exception as e:
                messagebox.showerror("Error", str(e))



    def insert_customer_details(self, current_date_time):
        try:
            conn = mysql.connector.connect(
                host="localhost",
                user="root",
                password="Kpml@11009",
                database="python"
            )
            cursor = conn.cursor()
    
            # Get the current date and time
            # current_date_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
            query = "INSERT INTO Customerdetails (Bill_No, Customer_Name, Mobile_No, Email, Date_Time) VALUES (%s, %s, %s, %s, %s)"
            values = (
                self.Bill_no.get(),
                self.Customer_name.get(),
                self.Mobile_no.get(),
                self.Email.get(),
                current_date_time  # Add date and time here
            )
    
            cursor.execute(query, values)
            conn.commit()
            conn.close()
            messagebox.showinfo("Saved", f"Customer details for Bill No: {self.Bill_no.get()} saved successfully")
        except Exception as e:
            messagebox.showerror("Error", str(e))




if __name__ == '__main__':
    root = Tk()
    obj = Bill_App(root)
    root.mainloop()
```
